```python
# Importing Packages
import matplotlib.pyplot as plt
import random

# Creating Coin Function
def flip_coin():
    coin_1 = random.randint(1, 2)
    coin_2 = random.randint(1, 2)

    # Determining if both coins have the same value
    if coin_1 == coin_2:
        same_num = True
    else:
        same_num = False
    return same_num

def presents():
  print("Our simple game will involve two coins. In order to win, the player needs to roll the same number on both dice.")

# Inputs
num_simulations = 1000
max_num_rolls = 100
bet = 1

# Tracking
win_probability = []
end_balance = []

```

```python
# For loop to run for the number of simulations desired
for i in range(num_simulations):
    balance = [1000]
    num_flip = [0]
    num_wins = 0    # Run until the player has rolled 1,000 times
    while num_flip[-1] < max_num_rolls:
        same = flip_coin()        # Result if the dice are the same number
        if same:
            balance.append(balance[-1] + 4 * bet)
            num_wins += 1
        # Result if the dice are different numbers
        else:
            balance.append(balance[-1] - bet)

        num_flip.append(num_flip[-1] + 1)# Store tracking variables and add line to figure
    win_probability.append(num_wins/num_flip[-1])
    end_balance.append(balance[-1])
    plt.plot(num_flip, balance)
```

```python
# Averaging win probability and end balance
overall_win_probability = sum(win_probability)/len(win_probability)
overall_end_balance = sum(end_balance)/len(end_balance)# Displaying the averages
print("Average win probability after " + str(num_simulations) + " runs: " + str(overall_win_probability))
print("Average ending balance after " + str(num_simulations) + " runs: $" + str(overall_end_balance))
```
